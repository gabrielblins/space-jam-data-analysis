```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

X = player_df.dropna(how='any').iloc[:,5:-1].to_numpy()
y = player_df.dropna(how='any').iloc[:,-1]

linear_regression = LinearRegression()

scores = cross_val_score(linear_regression, X, y, cv=10, scoring='r2')

entries, bin_edges, patches = plt.hist(scores,bins=10)

print('Mean r2 score: {:.4f}'.format(scores.mean()))
for i in range(len(entries)):
    if entries[i] > 0:
        print('{:.0f}% of r2 scores are between {:.4f} and {:.4f}'.format(entries[i]*100/len(entries), bin_edges[i],bin_edges[i+1]))
```

```python
linear_regression.fit(X,y)
```

```python
mask = player_df.isnull().any(axis=1)
mask
```

## Modulo 2 - Usando os dados para analisar partidas

```python
import pandas as pd
```

```python
player_df_final = pd.read_csv('player_data_final.csv')
```

```python
ts_df = player_df_final.iloc[26:, :]
ts_df
```

```python
ts_name_df = pd.read_csv('tune_squad.csv', sep='\t')
ts_name_df
```

```python
ts_df = pd.merge(ts_df, ts_name_df, on='ID', how='left', suffixes=('_type', '_name'))
ts_df.head()
```

```python
column_list = list(ts_df)

player_name = column_list.pop()
column_list[1] = player_name

ts_df = ts_df[column_list]
ts_df.head()
```

```python
game_stats_cols = list(ts_df.iloc[:,7:-1])
game_stats_stdevs = []

for stat in game_stats_cols:
    game_stats_stdevs.append(ts_df[stat].std())

stats_std = pd.Series(game_stats_stdevs, index=game_stats_cols)
stats_std
```

```python

from sklearn.linear_model import LinearRegression
from numpy.random import randn

X = player_df_final.iloc[:,7:-1].to_numpy()
y = player_df_final['PER']
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
print('Iterao # \thigh PER \tlow PER')

for i in range(10):

    game_df = pd.DataFrame(columns=game_stats_cols,index=list(ts_df['player_name']))

    for stat in game_stats_cols:

        game_df[stat] = list(ts_df[stat] + randn(len(ts_df)) * stats_std[stat])

    game_df['PER'] = lin_reg.predict(game_df)

    print('Iterao {}'.format(i+1) + ' \t' + game_df['PER'].idxmax() + ' \t' + game_df['PER'].idxmin())
```

```python
number_of_iterations = 4
df_list = [pd.DataFrame(columns=game_stats_cols,index=list(ts_df['player_name'])) for i in range(number_of_iterations)]

for df in df_list:
    for stat in game_stats_cols:
        df[stat] = list(ts_df[stat] + randn(len(ts_df)) * stats_std[stat])
    df['PER'] = lin_reg.predict(df)
game_df = pd.concat(df_list)
game_df.rename_axis('player_name', inplace=True)

minutes = [(x//len(ts_df))*12 for x in range(len(game_df))]
game_df['minutes'] = minutes
game_df.set_index('minutes', append=True, inplace=True)
game_df = game_df.swaplevel()

game_df
```

```python
game_df.to_csv('game_stats.csv')
```
